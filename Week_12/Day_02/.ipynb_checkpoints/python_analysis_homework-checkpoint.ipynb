### 3. Create new dataframe

```python
drinks = drinks_content.loc[:,["Beverage_category", "Beverage", "Beverage_prep", "Calories"]].copy()

drinks
```

```python
# check for missing values
drinks.isnull().sum()
```

### 4. Create `calorie_diff`

```python
drinks["calorie_diff"] = drinks["Calories"] - 135

drinks
```

### 5. Find mean calories in each category

```python
drinks.groupby("Beverage_category").mean("Calories")
```

### 6. Find `beverage_prep` type with more than average calories

```python
# Create variable which calculates and stores mean `Calories` value across all drinks
mean_calories = drinks["Calories"].mean()

mean_calories
```

```python
# Create new df which is the average `Calories` per `Beverage_prep`
calories_per_prep = drinks.groupby("Beverage_prep").mean("Calories")

calories_per_prep
```

```python
# Reset index after creating summary
calories_per_prep = calories_per_prep.reset_index()

calories_per_prep
```

```python
# filter new grouped df with stored `mean_calories` value
calories_per_prep[calories_per_prep["Calories"] > mean_calories]
```

### 7. Lowest calorie `coffee` drink

```python
# check beverage categories
list(drinks["Beverage_category"].unique())
# I would assume that everything but "Tazo Tea Drinks" and "Smoothies" could be coffee
```

```python
# filter `drinks` where beverage category is **not** Tazo tea or Smoothies

coffee_drinks = drinks[~drinks["Beverage_category"].isin(["Tazo® Tea Drinks", "Smoothies"])]

coffee_drinks
```

```python
coffee_drinks = coffee_drinks.reset_index()
```

```python
# sort coffee drinks by ascending calories, get the lowest
coffee_drinks.sort_values("Calories").head(1)

# the coffee drink with the lowest calories is Brewed Coffee which has 3 calories
```

### 8. Create `espresso_drinks` 

```python
# filter drinks for "Classic Espresso Drinks"
espresso_drinks = drinks[drinks["Beverage_category"].isin(["Classic Espresso Drinks"])]

espresso_drinks
```

### 9. Find mean `Calories` per `Beverage prep`

```python
espresso_drinks_bev_prep = espresso_drinks.groupby("Beverage_prep")["Calories"].mean()

espresso_drinks_bev_prep
```

### 10. Sort `Calories` in descending order

```python
espresso_drinks_bev_prep = espresso_drinks_bev_prep.reset_index()
espresso_drinks_bev_prep.loc[:, ["Beverage_prep", "Calories"]].copy()


espresso_drinks_bev_prep.sort_values("Calories", ascending= False)
```

### Extension

```python
drinks
```

```python
#average calories per bev category

calories_per_category = drinks.groupby("Beverage_category")["Calories"].mean()
calories_per_category = calories_per_category.reset_index()
calories_per_category.loc[:, ["Beverage_category", "Calories"]].copy()


calories_per_category
```

```python
sns.barplot(data = calories_per_category, x = "Calories", y = "Beverage_category", order=calories_per_category.sort_values("Calories").Beverage_category)
plt.xlabel("Average Calories")
plt.ylabel("Beverage Category")
plt.title("Average Calories per Beverage Category")
```

The above plot shows that smoothies on average have the highest calories and coffees have the lowest.
